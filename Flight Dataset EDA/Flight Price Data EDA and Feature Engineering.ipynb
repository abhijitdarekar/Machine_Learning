The column(Feature) : Date_of_Journey is to be broken to Day, Month , Year columns


```python
#Feature Engineering
#Diving the date_journey to DD , MM , YYYY
df_final['Date']=df_final['Date_of_Journey'].str.split('/').str[0]
df_final['Month']=df_final['Date_of_Journey'].str.split('/').str[1]
df_final['Year']=df_final['Date_of_Journey'].str.split('/').str[2]
```

```python
df_final.head()
```

```python
df_final['Date']=df_final['Date'].astype(int)
df_final['Month']=df_final['Month'].astype(int)
df_final['Year']=df_final['Year'].astype(int)
```

```python
df_final.info()
```

```python
df_final.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df_final.head()
```

```python
#Feature Engineering on Arrival_Time Feature

df_final['Arrival_Time'].str.split(' ').str[0]
```

```python
df_final['Arrival_Time']=df_final['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
df_final.head()
```

```python
df_final['Arrival_hour']=df_final['Arrival_Time'].str.split(':').str[0]
df_final['Arrival_min']=df_final['Arrival_Time'].str.split(':').str[1]
```

```python
df_final.head()
```

```python
df_final['Arrival_min']=df_final['Arrival_min'].astype(int)
df_final['Arrival_hour']=df_final['Arrival_hour'].astype(int)
```

```python
df_final.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df_final.head()
```

```python
#Feature Engineering for Departure Time Similar to Time

df_final['Dep_hour']=df_final['Dep_Time'].str.split(':').str[0]
df_final['Dep_min']=df_final['Dep_Time'].str.split(':').str[1]


df_final['Dep_hour']=df_final['Dep_hour'].astype(int)
df_final['Dep_min']=df_final['Dep_min'].astype(int)

df_final.drop('Dep_Time',axis=1,inplace=True)
```

```python
df_final.head()
```

```python
df_final.head()
```

```python
# Dealing with Route

# Total numbe rof unique stops to decide whether removing of route will be good enough or not 
df_final.Total_Stops.unique()
```

```python
df_final.Total_Stops=df_final.Total_Stops.map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
# Dealing with NAN in route Section and Total_stops 

df_final[df_final.Total_Stops.isnull()]
```

```python
#Dropping the ROute as Route can be calaculated by total_stops 

df_final.drop('Route',axis=1,inplace=True)
```

```python
df_final.head()
```

```python
# Feature Dealign with Additional Info

df_final.Additional_Info.unique()
```

```python
df_final['duration_hour']=df_final['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df_final.drop(6474,axis=0,inplace=True)
df_final.drop(2660,axis=0,inplace=True)
```

```python
df_final['duration_hour']=df_final['duration_hour'].astype('int')
```

```python
df_final.drop('Duration',axis=1,inplace=True)
```

```python
df_final.head()
```

```python
# Finding unique Airlines

df_final.Airline.unique()
```

```python
df_final.Airline.value_counts().reset_index()
```

```python
# Encoding all the Catogerical values usning lable encoder


from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df_final['Airline']=labelencoder.fit_transform(df_final['Airline'])
df_final['Source']=labelencoder.fit_transform(df_final['Source'])
df_final['Destination']=labelencoder.fit_transform(df_final['Destination'])
df_final['Additional_Info']=labelencoder.fit_transform(df_final['Additional_Info'])
```
